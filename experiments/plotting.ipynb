### Plotting Notebook
The following is a plotting notebook made to plot multiple streams of architecture data at once for comparison purposes.

The dataloading primitive is as follows: 

1. Provide a txt file containing plain formatted lists (`See vgg16_v1_performance.txt for example`)

2. Parse data using `read_data_from_file` for each file storing retrieved string into a list

3. Pass entire list of datasets to `plot_all_metrics` for simultaneous plotting

#### VGG_16

```python

file_path_1= '../performance_data/vgg_16/model_1/model_1.txt'  
data_1 = read_data_from_file(file_path_1)
file_path_2= '../performance_data/vgg_16/model_2/model_2.txt'  
data_2 = read_data_from_file(file_path_2)
file_path_3= '../performance_data/vgg_16/model_3/model_3.txt'  
data_3 = read_data_from_file(file_path_3)
file_path_4= '../performance_data/vgg_16/model_4/model_4.txt'  
data_4 = read_data_from_file(file_path_4)
data = [data_1,data_2,data_3,data_4]
labels= ["lr:1e-3/SGD","lr:1e-3/Adam","lr:2e-5/SGD","lr:2e-5/Adam"]
plot_all_metrics(data,labels,title="vgg_16")
```
